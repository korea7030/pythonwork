### Perceptron 으로 AND 회로 구현(1)

```python
def AND(x1, x2):
    w1,w2,theta = 0.5,0.5,0.7 # 임의의 값 지정(w : 가중치 / theta : 임계치)
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
```

```python
AND(1,1)
```

### Perceptron 으로 AND 회로 구현(2) - 가중치와 편향 추가

```python
import numpy as np

def AND_2(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])  # w : 입력값에 대한 영향력을 조절
    b = -0.7  # 편향 : 활성화 하느냐 조정하는 변수
    
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND_2(0,0)
```

```python
AND_2(1,1)
```

### NAND, OR 회로 구현

```python
Image.open('OR_graph.png')
```

```python
def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5]) # AND와 가중치가 반대
    b = 0.7 # AND와 편향 반대
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
    
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x)+b
    
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
NAND(0,0)
```

```python
OR(0,1)
```

###### 하나의 퍼셉트론으로는 구현이 불가능하므로, 층을 더 쌓아 다중 퍼셉트론으로 만들어야 한다.

## 기존 게이트로 조합(AND + NAND + OR)

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    
    return y
```

```python
XOR(0,0)
```
