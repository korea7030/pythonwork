```python
import numpy as np
import pandas as pd
import psycopg2
import datetime
```

```python
conn = psycopg2.connect("host='localhost' dbname='db_book' user='postgres' password='1234'")
cur = conn.cursor()
cur.execute("SELECT * from book_info")
raw_data = cur.fetchall()
```

```python
col_list = ['book_sequence', 'book_title','book_subtitle', 'author', 'translator', 'publisher', 'publisher_date', 'book_format', 'pages', 'rating', 'cate_cd', 'start_date', 'end_date', 'reg_date', 'book_essay_url', 'poster_url', 'read_status', 'borrowed_yn', 'book_essay']
```

```python
raw_data
```

```python
## list를 pandas dataframe으로 변환
raw_df_data = pd.DataFrame(raw_data, columns = col_list)
```

```python
raw_df_data.head()
```

```python
## 읽은일자 변수 추가
raw_df_data['readed_day'] = raw_df_data.apply(lambda row: (row.end_date - row.start_date).days, axis=1)
```

```python
from datetime import timedelta
def get_weekends(start_date, end_date):
    week_yn = 0
    st = start_date
    et = end_date
    
    all_days = [start_date + timedelta(days=x) for x in range((end_date-start_date).days + 1)]
    week_days = sum(1 for d in all_days if d.weekday() > 5)
    
        
    return week_days
```

```python
# 주말일수 변수 추가
raw_df_data['week_days_count'] = raw_df_data[["start_date", "end_date"]].apply(lambda row: get_weekends(row['start_date'], row['end_date']), axis=1)
```

```python
raw_df_data.head()
```

```python
raw_df_data['borrowed_yn'].replace('Y', 1, inplace=True)
raw_df_data['borrowed_yn'].replace('N', 0, inplace=True)
```

```python
raw_df_data['book_format'].replace('eBook', 1, inplace=True)
raw_df_data['book_format'].replace('Book', 0, inplace=True)
```

```python
raw_df_data.head()
```

```python
data_x, data_y = raw_df_data.drop(['book_sequence','cate_cd','book_title','book_subtitle','author', 'readed_day', 'translator', 'publisher', 'publisher_date', 'reg_date', 'book_essay_url', 'poster_url', 'book_essay', 'read_status', 'start_date', 'end_date'], axis=1), raw_df_data['readed_day']
```

```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=5)
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
diabetes_y_pred = lr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', lr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, diabetes_y_pred))
```

```python
lr.coef_[1]
```

#### model에 대한 결과 정보 

```python
from scipy import stats

params = np.append(lr.intercept_, lr.coef_)
predictions = lr.predict(X_test)

MSE = (sum((y_test-predictions)**2))/(len(X_test)-len(X_test.columns))

var_b = MSE*(np.linalg.inv(np.dot(X_test.T, X_test)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/sd_b

p_values = [2*(1-stats.t.cdf(np.abs(i), (len(X_test)-1))) for i in ts_b]

sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

regression_df = pd.DataFrame()
regression_df["Coefficients"],regression_df["Standard Errors"],regression_df["t values"], regression_df["Probabilites"] = [params,sd_b,ts_b,p_values]
print(regression_df)
```
