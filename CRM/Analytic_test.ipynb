```python
customer = pd.read_csv('A항공사 고객프로파일.txt', sep=",", encoding = 'utf-8')
```

```python
customer.head()
```

```python
history = pd.read_csv('A항공사 구매이력.txt', sep=",", encoding = 'utf-8')
```

```python
history.head()
```

```python
merge_data = pd.merge(customer,history, on = 'NO')
```

```python
merge_data.head()
```

```python
merge_data.describe()
```

```python
merge_data['회원등급'].value_counts()
```

```python
merge_data['일자'] = pd.to_datetime(merge_data['일자'], format='%Y%m%d')
```

```python
merge_data['발권일'] = pd.to_datetime(merge_data['발권일'], format='%d-%m-%Y')
```

```python
merge_data.info()
```

```python
merge_data['발권연도'] = merge_data.발권일.dt.year
```

```python
merge_data.head()
```

```python
merge_data['NET'].groupby([merge_data['회원등급'], merge_data['발권연도']]).sum()
```

```python
grade_ticketing_net = merge_data['NET'].groupby([merge_data['회원등급'], merge_data['발권연도']]).sum()
```

```python
grade_ticketing_net
```

```python
grade_ticketing_net_df = grade_ticketing_net.to_frame()
```

```python
grade_ticketing_net_df = grade_ticketing_net_df.reset_index()
```

```python
grade_ticketing_net_df
```

```python
grade_ticketing = merge_data['NET'].groupby(merge_data['회원등급']).sum()
```

```python
grade_ticketing_df = grade_ticketing.to_frame()
```

```python
grade_ticketing_df.columns = ['NET_SUM']
```

```python
grade_ticketing_df = grade_ticketing_df.reset_index()
```

```python
grade_ticketing_df
```

```python
grdae_ticketing_net_mg = pd.merge(grade_ticketing_net_df,grade_ticketing_df, on = "회원등급")
```

```python
grdae_ticketing_net_mg
```

```python
grdae_ticketing_net_mg['운임료사용비율'] = (grdae_ticketing_net_mg['NET'] / grdae_ticketing_net_mg['NET_SUM'])*100
```

```python
grdae_ticketing_net_mg.sort_index(by=['회원등급','운임료사용비율'], ascending = [True, False])
```
