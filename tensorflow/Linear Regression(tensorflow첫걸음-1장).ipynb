##### y = 0.1x + 0.3 관계를 가지는 데이터에 대한 plot

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

num_points = 1000
vectors_set = []

for i in range(num_points):
    x1 = np.random.normal(0.0, 0.55)
    y1 = x1*0.1 + 0.3+np.random.normal(0.0, 0.03)
    vectors_set.append([x1,y1])
    
x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]

plt.plot(x_data,y_data, 'ro')
plt.legend()
plt.show()
```

#### tensorflow 사용

```python
import tensorflow as tf

W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
# b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
y = W*x_data+b # Hypothesis

loss = tf.reduce_mean(tf.square(y-y_data)) # 가설값 - 실제 데이터의 제곱
optimizer = tf.train.GradientDescentOptimizer(0.5) # learning rate : 0. 
train = optimizer.minimize(loss)

# 알고리즘 실행
init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)

# gradient descent 그리기 위한 cost-W 그래프(그리려 했으나.. 실패)
# w_data = []
# cost_data = []

for step in range(8):
    sess.run(train)
    print(step, sess.run(loss), sess.run(W)) ## 비용함수 감소 확인
    # w_data.append(sess.run(W))
    # cost_data.append(sess.run(loss))
    
print (sess.run(W), sess.run(b))

# gradient descent graph 그리기
# plt.plot(w_data, cost_data,'ro')
# plt.legend()
# plt.ylabel('cost')
# plt.xlabel('W')
# plt.show()

## 그래프 출력
plt.plot(x_data, y_data, 'ro')
plt.plot(x_data, sess.run(W)*x_data+sess.run(b))
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
